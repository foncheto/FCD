**ACTIVIDAD 5**

Volvamos a cargar las datas utilizadas "comunas" y "datos_comunas_mod" y las juntamos en una solo dataframe.

Veamos el número de datos faltantes por variable y eliminamos estas filas y verifiquemos que nuestra nueva Data no tiene datos faltantes

```python
df_comunas1 = pd.read_csv("datos/comunas.csv")
df_comunas1
```

```python
df_comunas2=pd.read_csv("datos/datos_comunas_mod.csv")
df_comunas2
```

```python
df_juntas = pd.merge(df_comunas1,df_comunas2, how="outer", left_on="codigo",right_on="id")
df_juntas
```

```python
df_juntas.isnull().sum()
```

```python
df1 = df_juntas.dropna()
df1
```

```python
df1.isnull().sum()
```

Tambien podriamos eliminar la columna entera que contiene datos faltantes:

```python
dfsc =df_comunas2.dropna(axis =1)
dfsc
```
