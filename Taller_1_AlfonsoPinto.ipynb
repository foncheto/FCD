**ACTIVIDAD 7:**

Se define una nueva variable como Indice que corresponde al producto
entre Defense y Attack

Se desea agregar una nueva variable Defense\_10 que corresponde a
Defense multiplicado por 10, defina el codigo y ejecútelo

```python
#para esta actividad vamos a definir una función,
#que consiste en multiplicar por 10 una determinada fila de un dataframe

def Defence_n(col):
  result=col*10
  return result
df['Defense_10']=df['Defense'].apply(Defence_n)
df
```

Para utilizar la función anterior, sobre una variable "var" de un dataframe "df", para formar una nueva variable llamada "new_var", puede realizar lo siguiente:

df["new_var"] = df["var"].apply(Defence_n)

el resultado de la operación anterior es un dataframe que contiene la información del dataframe **df** junto con la nueva columna "new_var", que proviene del resultado de multiplicar por 10 la columna "var"
