**Encontremos un Intervalo de Confianza utilizando la distribución t tamaño de muestra menor de 30**

```python
data = [12, 12, 13, 13, 15, 16, 17, 22, 23, 25, 26, 27, 28, 28, 29]
```

```python
stats.t.interval(0.95, df = len(data)-1, loc = np.mean(data), scale = stats.sem(data))
```

```python
import numpy as np
import scipy.stats as st
fb_data = np.random.randint(15, 20, 80)
# create 90% confidence interval
resultado=st.norm.interval(0.90,
                 loc=np.mean(fb_data),
                 scale=st.sem(fb_data))
print(resultado)
```

**Realicemos una prueba de hipotesis para la actividad planteada de la embotelladora de bebidas**


```python
volumes =[220.1, 220.5, 221.2, 221.8, 222.5, 223.1, 223.7]
result = stats.ttest_1samp(volumes,225, alternative = 'less')
print(result)
print("-----")
print(round(result[1],7))
```

**¿Que significa este resultado?**

p- value: 0,00038 es menor a 0,05. Por lo que se rechaza la hipotesis nula, es decir, que el contenido es menor al especificado.
