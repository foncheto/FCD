**Intervalo de confianza**

Un ingeniero de control de calidad midió el espesor de la pared de 25 botellas de vidrio de dos litros. 
Encontrar un IC del 90 % para la media del espesor de la pared de las botellas.
Los valores medidos son los siguientes:
mediciones 

a=[4.1,4,4.2,4.1,4.2,4.005,4.05,4,3.8,4.1,4,3.9,4.05,3.9,4.2,4.05,4.05,4.05,4.07,4.07,4.05,4.3,4,3.8,4.2]

```python
a=[4.1,4,4.2,4.1,4.2,4.005,4.05,4,3.8,4.1,4,3.9,4.05,3.9,4.2,4.05,4.05,4.05,4.07,4.07,4.05,4.3,4,3.8,4.2]
```

```python
## Utilice stats.t.interval visto la clase pasada 
stats.t.interval(0.90, df = len(a)-1, loc = np.mean(a), scale = stats.sem(a))
```

Se espera que la media del espesor de la botella esté considerado en el intervalo obtenido con un 90% de confianza.

**Explique que significa este resultado**

**Por otro lado, la empresa que vendió las botellas le indicó que estas tenian un espesor uniforme de 4,05.**

Valide si estas botellas cumplen con la promesa del vendedor.

- Indique claramente Ho y H1
- Indique el p-value obtenido
- Explique el resultado

Ho=Distribuye normal 4,05

H1=No distribuye normal 4,05

```python
a=[4.1,4,4.2,4.1,4.2,4.005,4.05,4,3.8,4.1,4,3.9,4.05,3.9,4.2,4.05,4.05,4.05,4.07,4.07,4.05,4.3,4,3.8,4.2]
result=stats.ttest_1samp(a,4.05,alternative='less')
print(result)
print('------')
print(round(result[1],7))


```

Con el p value anterior podemos concluir que la hipotesis nula no se rechaza debido a que nuestro nivel de confianza (0.1, 90%) es menor a nuestro p value obtenido. Por lo tanto H0 (Distribuye normal 4,05) no se rechaza.
