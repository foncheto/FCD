```python
df_outer = pd.merge(comunas1,comunas2, how="outer", left_on="codigo",right_on="id")
df_outer
```

```python
df_inner = pd.merge(comunas1,comunas2, how="inner", left_on="codigo",right_on="id")
df_inner
```

```python
df_left = pd.merge(comunas1,comunas2, how="left", left_on="codigo",right_on="id")
df_left
```

```python
df_right = pd.merge(comunas1,comunas2, how="right", left_on="codigo",right_on="id")
df_right
```
